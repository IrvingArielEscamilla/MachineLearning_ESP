# Feedback de tus modelos

Ciertos modelos como la regresion o los arboles se dicen "Interpetables".Esto quiere decir que  de los resultados de los modelos podemos sacar conclusiones o "insights"

En particular la regresion Lasso es interpetable:
Mientras mas grande el coeficiente para feature, mas relevante es esta para la regresion.
La regresion Lasso trata de seleccionar un pequenio numero de features relevantes.

```python
X = pd.read_csv('/home/algoritmia/Dropbox/ml_intro/vol/intermediate_results/X.csv')
y = X['worldwide_gross']
X = X.drop('worldwide_gross',axis=1)
```

```python
from sklearn.linear_model import Lasso
model = Lasso()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
len(X_train)/len(X)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#parametros de la regresion lineal
model.coef_
```

```python
 var  = np.floor(np.log10(np.abs(model.coef_)))
```

```python
plt.rcParams["figure.figsize"] = [12,8]
plt.plot(var)
plt.xticks(np.arange(7),list(X.columns));
```

# Metodos de seleccion automatica de features

Skalearn posee una serie de metodos para selecionar las mejores features. Estos metodos
los puedes encontrar en sklearn.feature_selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression

selector = SelectKBest(mutual_info_regression, k=4)
selector.fit(X,y)
```

```python
scores = selector.scores_
plt.rcParams["figure.figsize"] = [12,8]
plt.plot(scores)
plt.xticks(np.arange(7),list(X.columns));
```

  Del analisis univariante obtenemos que las mejores features son:
  production_budget
  cast_total_faceebook_likes
  budget
